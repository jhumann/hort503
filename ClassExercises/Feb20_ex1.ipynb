```python
import pandas as pd
import numpy as np

Poplar_annot = pd.read_table('Ptrichocarpa_444_v3.1.annotation_info.txt')

Poplar_annot.head(5)
```

```python
GO_terms = pd.read_table('GO.terms.txt')
GO_terms.columns = ['Vocabulary', 'GO', 'Definition']
GO_terms.head()
```

```python
Go_sort = Poplar_annot[['peptideName', 'GO']]
#Go_sort.head()
No_Nan = Go_sort.dropna()
#No_Nan.head()

GO_dfs = []

for index, row in No_Nan.iterrows():
    terms = row.GO.split(',')
    protein = np.repeat(row.peptideName, len(terms))
    df = pd.DataFrame({'peptideName': protein, 'GO': terms})
    GO_dfs.append(df)


poplar_go = pd.concat(GO_dfs)
poplar_go.head()
```

```python
merged = poplar_go.set_index('GO').join(GO_terms.set_index('GO'), how='inner')
merged.head()
```
