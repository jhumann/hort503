```python
import pandas as pd
PI = pd.read_table('PI_DataSet.txt')
PI.head()
```

```python
aa_data = PI.drop(['FPV', 'ATV', 'IDV', 'LPV', 'NFV', 'SQV', 'TPV', 'DRV'], axis=1)

aa_data = pd.melt(aa_data, id_vars=['SeqID'], var_name='AA_pos_raw', value_name='AA')

aa_data["AA_pos"] = aa_data["AA_pos_raw"].str.extract('(\d+)', expand=False).astype(int)
aa_data = aa_data[["SeqID", "AA_pos", "AA"]]
aa_data = aa_data.sort_values(['SeqID', 'AA_pos'])
aa_data = aa_data.reset_index(drop=True)
aa_data.head()
```
