```python
import pandas as pd

corn = pd.read_csv('corn.csv')
corn.head()
```

```python
Bt_only = corn.iloc[3:17, 0:19]

Bt_only.columns = corn.iloc[1,0:19]
Bt_only = Bt_only.reset_index(drop=True)
Bt_only
```

```python
state = [0]
state.extend(list(range(19,37)))
herb_tol = corn.iloc[3:17, state]

herb_tol.columns = corn.iloc[1,state]
herb_tol = herb_tol.reset_index(drop=True)
herb_tol
```

```python
Bt_tidy = pd.melt(Bt_only, id_vars=['State'], var_name='Year', value_name='Value')
Bt_tidy['Crop'] = 'corn'
Bt_tidy['Crop Title'] = 'Genetically engineered (GE) corn'
Bt_tidy['Variety'] = 'Insect-resistant (Bt) only'
Bt_tidy['Unit'] = 'Percent of all corn planted'
Bt_tidy
```

```python
herb_tidy = pd.melt(herb_tol, id_vars=['State'], var_name='Year', value_name='Value')
herb_tidy['Crop Title'] = 'Genetically engineered (GE) corn'
herb_tidy['Variety'] = 'Herbicide-tolerant only'
herb_tidy['Unit'] = 'Percent of all corn planted'
herb_tidy
```

```python
corn_tidy = pd.concat([Bt_tidy, herb_tidy])
corn_tidy
```

```python
corn_tidy = corn_tidy.dropna()
corn_tidy['Year'] = corn_tidy['Year'].astype('int')
corn_tidy['Value'] =corn_tidy['Value'].astype('int')
corn_tidy.head()
```

```python
import matplotlib 
%matplotlib inline

State = 'Iowa'
Variety = 'Insect-resistant (Bt) only'
xticks = corn_tidy['Year'].drop_duplicates().sort_values()
corn_tidy[(corn_tidy['State'] == State) & (corn_tidy['Variety'] == Variety)].plot(
    kind='line', x='Year', y='Value',
    xticks=xticks, rot=45,
    title='Percentage of Insect-resistant (Bt) only GE corn in Iowa')
```

```python
cit = corn_tidy[corn_tidy['Variety'] == "Insect-resistant (Bt) only"]
citp = cit[['Year', 'State', 'Value']].pivot(index='Year', columns='State')
citp.head()
```

```python
citp.plot(kind='bar', rot=45, legend=False, figsize=(15,4),
         title='Percentage of Insect-resistant (Bt) only GE corn in all states')
```
